>  *Entrenamiento*



````python
pipe = Pipeline([('poly', PolynomialFeatures()), ('elastic', ElasticNet())])
pipe.set_params(poly__degree=2, elastic__alpha=5, elastic__l1_ratio=1, elastic__fit_intercept=True,elastic__normalize=True)

pipe.fit(XTransformed, y)
````


>  *Validación*




````python

from sklearn.model_selection import cross_val_score

scores=cross_val_score(pipe,XTransformed, y, scoring='r2' )
scores=np.mean(scores)
print(scores)
````
